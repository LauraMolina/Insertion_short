```python
from itertools import *
#import matplotlib.pyplot as plt

x = []
instru = []

def insertion_short(A):
    comparaciones = 0
    intercambios = 0
    for i in range(1,len(A)):
        temp = A[i]
        j = i
        comparaciones += 1
        while j > 0 and temp < A[j-1]:
            if j != i:
                comparaciones += 1
            A[j] = A[j-1]
            j = j-1
            intercambios += 1
        A[j] = temp
        if j > 0 and j != i:
            comparaciones += 1


    instrucciones = comparaciones + intercambios
    resultados = []
    resultados.append(instrucciones)
    resultados.append(comparaciones)
    resultados.append(intercambios)
    return resultados

for n in range(2,10):
    A = list(i for i in range(n))
    acum_ins = 0
    acum_c = 0
    acum_int = 0
    long = 0
    for a in permutations(A):
        acum = insertion_short(list(a))
        acum_ins += acum[0]
        acum_c += acum[1]
        acum_int += acum[2]
        long += 1

    print('ARREGLO DE TAMAÑO {}'.format(n))
    print('La cantidad de instrucciones promedio es: {}'.format(float(acum_ins / long)))
    print('La cantidad de comparaciones promedio es: {}'.format(float(acum_c / long)))
    print('La cantidad de intercambios promedio es: {}'.format(float(acum_int / long)))
    print('-----------------------')
    x.append(n)
    instru.append(float(acum_ins / long))

#plt.plot(x, instru)
#plt.xlabel('Tamaño arreglo')
#plt.ylabel('Cantidad de instrucciones promedio')
#plt.title('Instrucciones')
#plt.show()
```
